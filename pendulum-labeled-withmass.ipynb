```python
from matplotlib import pyplot as plt
from matplotlib import animation
from numpy import sin, cos, pi
import numpy as np
from scipy.integrate import odeint
from IPython.display import HTML

import matplotlib as mpl

mpl.rcParams['figure.dpi'] = 240

from matplotlib import rc

rc('text', usetex=True)
```

```python
def simple_pendulum_ode(conditions, t, =20, g=980):
    # state of  and  is in the conditions list
    ,  = conditions
    
    dydt = [, -g/*sin()]
    return dydt

_inital = pi/2 - 0.1
_inital = 0.0

t_list = np.linspace(0, 5, 1600)

 = 20

solutions = odeint(simple_pendulum_ode, [_inital, _inital], t_list)
# odeint solves for  and  over the interval in t_list, it pases conditions = [_inital, _inital] 
# for the first call of simple_pendulum_ode().

_solution_list = [i[0] for i in solutions]

x_0 = 0
y_0 = 40

y_list = [y_0 - *cos() for  in _solution_list]
x_list = [*sin() for  in _solution_list]

fig = plt.figure()


ax = fig.add_subplot(111, aspect='equal', autoscale_on=False,
                     xlim=(-50, 50), ylim=(0, 60))
plt.grid()

plt.title('Simple Pendulum on the Cartesian Plane')
plt.xlabel('$x$-axis (cm)')
plt.ylabel('$y$-axis (cm)')
plt.text(-13, 41, r'$$\frac{d^2\theta}{dt^2} + \frac{g}{\ell}\sin\theta = 0$$', fontsize = 13)

lns = []

for i in range(len(solutions)):
    
    line, = plt.plot([x_0, x_list[i]], [y_0, y_list[i]], color='k', lw=1)
    mass, = plt.plot([x_list[i]], [y_list[i]], 'ro', color = 'royalblue')
    
    time_text = plt.text(-41.2+9.8, 25+2, f'$t = {t_list[i]:.1f} s$')
    theta_text = plt.text(-44.5+10.3, 20+2, f'$$\\|\\theta\\| = {abs(_solution_list[i]):.2f}$$')
    sintheta_text = plt.text(-49+10.2, 15+2, f'$$\\|\\sin \\theta \\| = {abs(sin(_solution_list[i])):.2f}$$')
    
    
    lns.append([line, mass, time_text,theta_text,sintheta_text])
    


ani = animation.ArtistAnimation(fig, lns, interval=15)

HTML(ani.to_html5_video())
```

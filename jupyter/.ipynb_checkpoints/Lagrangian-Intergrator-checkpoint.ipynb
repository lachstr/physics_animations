```python
from scipy.misc import derivative
import scipy.integrate as integrate
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 1
g = 9.8

def x_0(t):
    return -g/2 * t**2 + 5*g*t
    
def v(t):
    return derivative(x, t)

def L(t):
    T = 0.5*m*(v(t))**2
    V = m*g*x(t)
    
    return T - V
```

```python
t = np.linspace(0, 10, 100)
plt.plot(t, x_0(t))
```

```python
S = integrate.quad(L, 0, 10)
```

```python
S
```

```python
def x_wrong(t):
    return 10*np.sin(2*t)

t = np.linspace(0, 10, 100)

def x(t, err_size=1):
    return x_0(t) + err_size*x_wrong(t)

plt.plot(t, x(t))
```

```python
S = integrate.quad(L, 0, 10)
S
```

### Animate here


```python
from matplotlib import rcParams, rc, animation
from IPython.display import HTML
##Warning this code requires a working latex installation to execute and ffmpeg


fig2 = plt.figure()
plt.title(f"")
plt.xlabel("t")
plt.ylabel("x(t)")

frame_count = 30

def err_size(frame, frame_count):
    return np.sin(np.pi/frame_count * frame)

ims = []
for frame in np.arange(frame_count):
    testx1 = plt.plot(t, x(t, err_size(frame, frame_count)), color='#1f77b4')
    S = plt.text(3, 20, f"{integrate.quad(L, 0, 10)}")
    
    ims.append(testx1, S)

im_ani = animation.ArtistAnimation(fig2, ims, interval=120,
                                   blit=False)

HTML(im_ani.to_html5_video())
```

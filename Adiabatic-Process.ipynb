```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('text', usetex=True)

plt.figure(num=None, figsize=(7, 5), dpi=150, facecolor='w', edgecolor='k')

def volume_f(pionpf, degrees_of_freedom):
    return pionpf**(degrees_of_freedom/(degrees_of_freedom+2))

from numpy import linspace

compression = linspace(0, 2, num=500)
final_volume_diatomic = volume_f(compression, degrees_of_freedom=5)
final_volume_monatomic = volume_f(compression, degrees_of_freedom=3)

plt.ylabel(r"Compression Factor $(V_f/V_i)$")
plt.xlabel(r"Pressure Ratio $(P_i/P_f)$")
diatomic_curve, = plt.plot(compression, final_volume_diatomic)

diatomic_curve.set_label("Diatomic Gas")

monatomic_curve, = plt.plot(compression, final_volume_monatomic)

monatomic_curve.set_label("Monatomic Gas")


split, = plt.plot([1,1],[0,1], '--')
split.set_label("Line of Compression/Expansion")

plt.text(1.25, 0.5, '$V_f/V_i = (P_i/P_f)^{\\frac{f}{f+2}}$')
plt.title("Adiabatic Compression Factors of Gases")
plt.legend()
plt.savefig("compressionfactors-plot.png")
```

```python
plt.figure(num=None, figsize=(7, 5), dpi=150, facecolor='w', edgecolor='k')


def delta_T(pionpf, degrees_of_freedom, initial_temp = 300):
    return ((pionpf)**(-2/(degrees_of_freedom+2)) - 1) * initial_temp

pressurespace = linspace(0.5, 1.5, num=500)

monatomic = delta_T(pressurespace, degrees_of_freedom=3)
diatomic = delta_T(pressurespace, degrees_of_freedom=5)

plt.xlabel(r"Pressure Ratio $(P_i/P_f)$")
plt.ylabel(r"Change in Temperature $($K$)$")


diatomic_curve, = plt.plot(pressurespace, diatomic)
monatomic_curve, = plt.plot(pressurespace, monatomic)

diatomic_curve.set_label("Diatomic Gas")

monatomic_curve.set_label("Monatomic Gas")

split, = plt.plot([1,1],[-50,1], '--')
split.set_label("Line of Compression/Expansion")

plt.text(1.1, 25, r'$$\Delta T = ((\frac{P_i}{P_f})^{\frac{-2}{f+2}} - 1)T_i$$')
plt.title('Temperature Change in Adiabatic Process at 300K')

plt.text(1.15, 15, r'Note: $\Delta W \propto \Delta T$')
plt.legend()

plt.savefig("temperaturechange-plot.png")
```

Case 1:

$$\Delta W=-\int_{V_i}^{V_i} P dV = 0 $$

As $\Delta W = 0  \implies \Delta U = \Delta Q$\
Equipartition of energy gives $\Delta U$ = $\frac{f}{2}Nk_B\Delta T$\
Ideal gas law $\Delta P = \frac{Nk_B\Delta T}{V}$


Case 2:

$$\Delta W=-\int_{V_i}^{V_f} P dV =-\int_{V_i}^{V_f} 0 dV =-0\int_{V_i}^{V_f} dV = 0 $$

As $\Delta W = 0  \implies \Delta U = \Delta Q$\
Equipartition of energy gives $\Delta U$ = $\frac{f}{2}Nk_B\Delta T$\
Ideal gas law $\Delta V = \frac{Nk_B\Delta T}{P}$

Case 3:

Equipartition of energy gives $\Delta U = 0$\
2nd law gives $\Delta W = -\Delta Q$

$$\Delta W=-\int_{V_i}^{V_f} P dV =-\int_{V_i}^{V_f} \frac{Nk_bT}{V} dV = -Nk_bT\int_{V_i}^{V_f} \frac{dV}{V} = -Nk_BT\ln{\frac{V_f}{V_i}} $$

So $\Delta W > 0 \implies V_f < V_i \implies \Delta V < 0$\
Similarly;
$\Delta W < 0 \implies V_f > V_i \implies \Delta V > 0$

By the ideal gas law under constant temperature\
$$PV=constant$$
So $\Delta V < 0 \implies \Delta P > 0$\
Similarly $\Delta V > 0 \implies \Delta P < 0$

Case 4:

By equipartition of energy $\Delta U$ = $\frac{f}{2}Nk_B\Delta T$

We know that (in quasistatic processes) $$\Delta T = ((\frac{P_i}{P_f})^{\frac{-2}{f+2}} - 1)T_i$$

So $\Delta T > 0 \implies P_f > P_i \implies \Delta P > 0$\
Similarly, $\Delta T < 0 \implies P_f < P_i \implies \Delta P < 0$

We know that $$PV^\frac{f+2}{f} = constant$$

$\Delta P > 0 \implies \Delta V < 0$\
$\Delta P < 0 \implies \Delta V > 0$


Case 5:

"An adiabatic process that is also quasistatic, the entropy is unchanged; such a process is called isentropic"

$$Q=TdS$$
$\Delta S = 0 \implies \Delta Q = 0$

By the 2nd law;
$\Delta U = \Delta W$

By the equipartition of energy;
$\Delta U$ = $\frac{f}{2}Nk_B\Delta T$

$$S = k_B\ln{\Omega}$$
In an ideal gas

$$\Omega(U,V,N)=f(N)V^NU^{3N/2}$$

$\Delta S = 0 \implies \Omega_i = \Omega_f$\
$V_i^N U_i^{3N/2} = V_f^N U_f^{3N/2}$\
$V_i/V_f = (U_f/U_i)^{N/2}$

So, $\Delta U > 0 \implies \Delta V < 0$\
$\Delta U < 0 \implies \Delta V > 0$

We know that in adiabatic processes $$PV^\frac{f+2}{f} = constant$$

$\Delta V < 0 \implies \Delta P > 0$\
$\Delta V > 0 \implies \Delta P < 0$
